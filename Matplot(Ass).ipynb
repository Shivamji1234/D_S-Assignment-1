```python
Matplotlib is a popular plotting library in Python used for creating a wide range of static, interactive, and animated visualizations. It provides a flexible and comprehensive set of functions for creating various types of plots, graphs, and charts. Matplotlib is commonly used for data visualization
1.Line Plot: Used to visualize the relationship between two variables over continuous data points, typically with lines connecting the data points.
2.Bar Plot: Used to compare different categories or groups by representing data with rectangular bars of varying heights or lengths.
3.Histogram: Used to display the distribution of a continuous dataset by dividing the data into bins and counting the number of occurrences in each bin.
4.Scatter Plot: Used to visualize the relationship between two variables, often with individual data points represented as dots on the plot.
5.Pie Chart: Used to represent data as a circle divided into sectors, where each sector represents a proportion of the whole data set.
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
Scatter Plot: Used to visualize the relationship between two variables, often with individual data points represented as dots on the plot.
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.scatter(x,y,c='r')
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It's a powerful tool for creating multiple plots in a single figure canvas, allowing you to display and compare multiple visualizations side by side.
```

```python
plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)
plt.show()
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.show()
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.show()
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

```python
Bar Plot: Used to compare different categories or groups by representing data with rectangular bars of varying heights or lengths.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,color='r')
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A box plot, also known as a box-and-whisker plot, is a graphical representation that provides a summary of the distribution and variability of a dataset. It displays the distribution of a dataset through its quartiles, central tendency, and outliers, making it useful for gaining insights into the data's spread and skewness.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(2,2,1)
plt.boxplot(box1)
plt.show()
plt.subplot(2,2,2)
plt.boxplot(box2)
plt.show()
```
