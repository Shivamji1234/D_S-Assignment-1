```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Seaborn is a powerful data visualization library in Python that is built on top of Matplotlib. It offers a higher-level interface for creating informative and attractive statistical graphics. Here are five types of plots that you can create using Seaborn, along with their uses:

1.Displot:
   - Use: same as Histogram used show count of distribution.
2. Scatter Plots:
   - Use: Scatter plots display the relationship between two continuous variables. They are useful for identifying patterns, clusters, and trends in the data.
   
3. Pair Plots:
   - Use: Pair plots are used to visualize pairwise relationships among multiple variables in a dataset. They create scatter plots for every combination of variables in the dataset.
   
4. Bar Plots:
   -Use: Bar plots are used to compare categorical data, such as the frequency of different categories or groups. They can visualize count, mean, or other aggregate values for each category.
   
5. Heatmaps:
   - Use: Heatmaps visualize a matrix-like dataset by assigning colors to the values in the matrix. They are useful for displaying correlations between variables or any kind of two-dimensional dataset.
   
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.head(3)
```

```python
sns.lineplot(x ="timepoint" , y = "signal",data=fmri,hue='event' )
```

```python
sns.lineplot(x ="timepoint" , y = "signal",data=fmri,hue='region' )
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.boxplot(x ='pclass', y = 'age',data=titanic)
```

```python
sns.boxplot(x ='pclass', y = 'fare',data=titanic)
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
diamond=sns.load_dataset("diamonds")
```

```python
diamond.head()
```

```python
sns.histplot(data=diamond,x='price',hue='cut',multiple='stack',weights=0.6,bins=10)
```
